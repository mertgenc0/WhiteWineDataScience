```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('winequality-white.csv', sep=';')
```

```python
dataset.head(20)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
# Graph of the link between alcohol rate and quality
plt.figure(figsize=(8,6))
df_grouped = dataset.groupby('quality')['alcohol'].mean()
df_grouped.plot(kind='line', marker='o')
plt.title('Average Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Alcohol')
plt.grid(True)
plt.show()
```

```python
dataset.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```
